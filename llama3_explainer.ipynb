## Identifying logits for target variables

Here we can observe that the input_ids contain 2 logits.

- The first logit is for <|begin_of_text|>
- The second logit is for the given target variable

```python
target_variables = ["Negative", "Neutral", "Positive"]
```

```python
target_logits = []

for target in target_variables:
    print("---" * 5, target, "---" * 5)

    # Encode the prompt
    token = tokenizer(target, return_tensors="pt")
    print(token)

    # Generate the logits
    decoded_token = tokenizer.decode(token.input_ids[0])
    print(decoded_token)

    # Extract the logit for the target token and append it to the list
    target_logits.append(token.input_ids[0][1:].tolist())
    print()

print("-" * 40)

for var, logit in zip(target_variables, target_logits):
    print(f"{var}: {logit}")
```

## Defining chat template and message wrappers

```python
def zero_shot_wrapper(text):
  return [
    {
       "role":"system",
       "content":"You are a sentiment classifier. Respond if the given text is Positive / Neutral / Negative / None"
    },
    {
       "role":"user",
       "content": text
    }
 ]
```

```python
def few_shot_wrapper(text):
  return [
   {
      "role":"system",
      "content":"You are a sentiment classifier. Respond if the given text is Positive / Neutral / Negative"
   },
   {
      "role":"user",
      "content":"I have had to take it to the geek squad guys to get it fixed because it has had viruses 3 times and i have not even had it for more then 4 months!"
   },
   {
      "role":"assistant",
      "content":"Negative"
   },
   {
      "role":"user",
      "content":"It certainly looks nice, but that nice dark blue top is a fingerprint magnet as others have pointed out."
   },
   {
      "role":"assistant",
      "content":"Negative"
   },
   {
      "role":"user",
      "content":"It is a big big, but since it has a 18.4\" screen what would you expect!"
   },
   {
      "role":"assistant",
      "content":"Neutral"
   },
   {
      "role":"user",
      "content":"I know once I upgrade the ram to 8GB it will pick up a lot."
   },
   {
      "role":"assistant",
      "content":"Neutral"
   },
   {
      "role":"user",
      "content":"Keyboard is nice and windows 8.1 is great, worried about that coming from xp."
   },
   {
      "role":"assistant",
      "content":"Positive"
   },
   {
      "role":"user",
      "content":"Its small enough where I can take it pretty much anywhere, but still has a big enough screen to get everything done."
   },
   {
      "role":"assistant",
      "content":"Positive"
   }
]
```

```python
def apply_chat_template(text, tokenizer, text_wrapper):
  messages = text_wrapper(text)

  input_ids = tokenizer.apply_chat_template(
      messages,
      add_generation_prompt=True,
      return_tensors="pt"
  ).to(model.device)

  return input_ids
```

### Invoking model to make prediction

```python
def make_prediction(input_ids):
    outputs = model.generate(
        example_input_ids,
        max_new_tokens=256,
        do_sample=True,
        temperature=0.6,
        top_p=0.9,
        eos_token_id=[
            tokenizer.eos_token_id,
            tokenizer.convert_tokens_to_ids("<|eot_id|>")
        ],
    )
    return outputs[0]

example_result = make_prediction(example_input_ids)
print('---------------------------------')
print(example_result)
```

### Checking top k predictions from the model

```python
def example_prediction(input_ids):
    model.eval()
    with torch.no_grad():
        output = model(input_ids=input_ids)

    last_token_logits = output.logits[0, -1, :]
    return torch.softmax(last_token_logits, dim=-1)
```

```python
example_predictions = example_prediction(example_input_ids)
print("ZERO SHOT EXAMPLE\n\nThe model has predicted total", len(example_predictions), "tokens.")

k = 10
print("Top", k, "predictions:")
top_predictions = example_predictions.topk(k)

predicted_tokens = tokenizer.convert_ids_to_tokens(top_predictions.indices, skip_special_tokens=True)

print("\nProbability |   Logit | Token")
print("------------|---------|-----------")
for token, ind, val in zip(predicted_tokens, top_predictions.indices, top_predictions.values):
    print(f"{val.item():11.4f} | {ind.item():7} | {token}")
```

```python
example_predictions_2 = example_prediction(apply_chat_template(example_text, tokenizer, few_shot_wrapper))
print("FEW SHOT EXAMPLE\n\nThe model has predicted total", len(example_predictions_2), "tokens.")

k = 10
print("Top", k, "predictions:")
top_prediction_2 = example_predictions_2.topk(k)

predicted_tokens_2 = tokenizer.convert_ids_to_tokens(top_prediction_2.indices)

print("\nProbability |   Logit | Token")
print("------------|---------|-----------")
for token, ind, val in zip(predicted_tokens_2, top_prediction_2.indices, top_prediction_2.values):
    print(f"{val.item():11.8f} | {ind.item():7} | {token}")
```

# Model Interpretability

```python
def get_functions(model, tokenizer, apply_chat_template, text_wrapper):
    def preprocess_text(text):
        tokens = tokenizer(text, return_tensors='pt')
        return tokenizer.decode(tokens['input_ids'][0], skip_special_tokens=True)

    def predict_proba(texts):
        preprocessed_texts = [preprocess_text(text) for text in texts]
        # print(preprocessed_texts)
        probabilities = []

        for text in preprocessed_texts:

            # Apply the chat template along with the wrapper
            input_ids = apply_chat_template(text, tokenizer, text_wrapper)

            # Make the prediction
            with torch.no_grad():
                outputs = model(input_ids=input_ids)
                last_token_logits = outputs.logits[0, -1, :]
                prob = torch.softmax(last_token_logits, dim=-1)
                prob = prob.cpu().numpy()

                # Extract the probabilities only for the target tokens
                prob = [prob[i] for i in target_logits]
                probabilities.append(prob)

        # print(probabilities)
        return probabilities

    return preprocess_text, predict_proba
```

# SHAP Explainer

```python
import shap

def execute_shap_explainer(explainer, text):
    # Generate SHAP values
    shap_values = explainer([text])

    # Visualize the explanation
    shap.text_plot(shap_values)

    return shap_values
```

## Zero-Shot Prompt Explainer

```python
_, predict_fn = get_functions(model, tokenizer, apply_chat_template, zero_shot_wrapper)
shap_explainer = shap.Explainer(predict_fn, tokenizer, output_names=target_variables)
```

```python
shap_values = execute_shap_explainer(shap_explainer, example_text)
```

```python
_2, predict_fn_2 = get_functions(model, tokenizer, apply_chat_template, few_shot_wrapper)
fs_shap_explainer = shap.Explainer(predict_fn_2, tokenizer, output_names=target_variables)
```

```python
shap_values = execute_shap_explainer(fs_shap_explainer, example_text)
```
